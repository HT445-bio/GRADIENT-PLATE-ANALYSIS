```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import levene

# Sample structured data
data = {
    'Antibiotic': ['MEC', 'AMP', 'CARB', 'CHLR', 'ERY'],
    'DBSCAN': [
        [0.3, 0.15, 0.29, 0.14, 0.5, 0.27, 0.4, 0.35, 0.38],
        [17.33, 16.41, 14.01, 16.55, 17.77, 15.97, 14.35, 16.11, 8.45, 10.5, 9.57, 9.9, 9.29, 8.78],
        [10.44, 12.16, 9.65, 11.3, 10.41, 12.29, 11.87, 11.24],
        [1.81, 1.87, 1.9, 1.76, 1.49, 2.16, 1.57, 1.71],
        [10.64, 14.99, 12.89, 20.88]
    ],
    'KNN': [
        [0.44, 0.67, 0.53, 0.58, 0.51, 0.6, 0.53, 0.58],
        [17.19, 17.48, 14.94, 15.28, 16.5, 18.46, 12.44, 12.59, 14.11, 14.31, 10.21, 14.45, 10.55, 9.91],
        [8.39, 11.66, 11.24, 11.16, 9.89, 12.24, 6.9, 11.06],
        [2.58, 2.68, 2.45, 3.65, 1.07, 1.71, 1.5, 1.6],
        [10.38, 14.36, 10.11, 13.23]
    ]
}

# Convert to tidy long-form DataFrame
rows = []
for ab, db_vals, knn_vals in zip(data['Antibiotic'], data['DBSCAN'], data['KNN']):
    rows.extend({'Antibiotic': ab, 'Method': 'DBSCAN', 'MIC': v} for v in db_vals)
    rows.extend({'Antibiotic': ab, 'Method': 'KNN', 'MIC': v} for v in knn_vals)

long_df = pd.DataFrame(rows)

# Plot and Levene's test for each antibiotic
for ab in long_df['Antibiotic'].unique():
    ab_df = long_df[long_df['Antibiotic'] == ab]
    db_vals = ab_df[ab_df['Method'] == 'DBSCAN']['MIC']
    knn_vals = ab_df[ab_df['Method'] == 'KNN']['MIC']

    # Levene's test for equal variances
    stat, p = levene(db_vals, knn_vals)

    # Plot
    plt.figure(figsize=(6, 5))
    sns.boxplot(data=ab_df, x='Method', y='MIC', palette='inferno')
    plt.title(f"{ab} MIC Comparison\nLevene’s Test p = {p:.4f}")
    plt.ylabel("MIC (µg/mL)")
    plt.tight_layout()
    plt.show()

    # Console output
    print(f"{ab} — Levene’s Test Result:")
    print(f"  Statistic: {stat:.4f}, p-value: {p:.4f}\n")

```

```python
import numpy as np
from scipy import stats

singles = {'AMP': [17.33, 16.41, 14.01, 16.55, 17.77, 15.97, 14.35, 16.11, 8.45, 10.5, 9.57, 9.9, 9.29, 8.78], 'CHLR': [1.81, 1.87, 1.9, 1.76, 1.49, 2.16, 1.57, 1.71]}
doubles = {'AMP+CHLR': [15.19, 11.46, 15.73, 14.99], 'CHLR+AMP':[3.9, 4.53, 3.43, 0.88]}
var_amp = np.var(singles['AMP'])
var_chlr = np.var(singles['CHLR'])
var_ampchlr = np.var(doubles['AMP+CHLR'])
var_chlramp = np.var(doubles['CHLR+AMP'])

# Compute the F-statistic
F_amp = var_amp / var_ampchlr if var_amp > var_ampchlr else var_ampchlr / var_amp
F_chlr = var_chlr / var_chlramp if var_chlr > var_chlramp else var_chlramp / var_chlr
# Degrees of freedom
df_amp = len('AMP') - 1
df_ampchlr = len('AMP+CHLR') - 1
df_chlr = len('CHLR') - 1
df_chlramp = len('CHLR+AMP') - 1

# Compute the p-value
# Two-tailed p-value
p_value_amp = 2 * min(stats.f.cdf(F_amp, df_amp, df_ampchlr), 1 - stats.f.cdf(F_amp, df_amp, df_ampchlr))
p_value_chlr = 2 * min(stats.f.cdf(F_chlr, df_chlr, df_chlramp), 1 - stats.f.cdf(F_chlr, df_chlr, df_chlramp))
print("Variance of AMP single:", var_amp)
print("Variance of AMP+CHLR:", var_ampchlr)
print("Variance of CHLR single:", var_chlr)
print("Variance of CHLR+AMP:", var_chlramp)
print("F-statistic AMP:", F_amp)
print("F-statistic CHLR:", F_chlr)
print("p-value AMP:", p_value_amp)
print("p-value CHLR", p_value_chlr)

print("IF p < 0.05, Variances are significantly different — use Welch's t-test.")
print("Else, Variances are not significantly different — use Student's t-test (equal variances).")



# AMP vs AMP/CHLR comparison
if p_value_amp < 0.05:
    stat, p_val = ttest_ind(singles['AMP'], doubles['AMP+CHLR'], equal_var=False)
    print(f"AMP vs AMP/CHLR — t-test: stat={stat:.3f}, p={p_val:.3f}")
else:
    stat, p_val = ttest_ind(singles['AMP'], doubles['AMP+CHLR'], equal_var=True)
    print(f"AMP vs AMP/CHLR — t-test: stat={stat:.3f}, p={p_val:.3f}")

#CHLR vs AMP/CHLR comparison
if p_value_chlr < 0.05:
    stat, p_val = ttest_ind(singles['CHLR'], doubles['CHLR+AMP'], equal_var=False)
    print(f"CHLR vs AMP/CHLR — t-test: stat={stat:.3f}, p={p_val:.3f}")
else:
    stat, p_val = ttest_ind(singles['CHLR'], doubles['CHLR+AMP'], equal_var=True)
    print(f"CHLR vs AMP/CHLR — t-test: stat={stat:.3f}, p={p_val:.3f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
antibiotics = ['Mecillinam', 'Ampicillin', 'Carbenicillin', 'Chloramphenicol',
               'Doxycycline', 'Erythromycin', 'Ciprofloxacin']
mic_data = {
    '96 Well Plate': [0.3, 0.15625, 12.5, 1.25, 1.61, 40, 0.016],
    'Gradient Plates': [0.3, 11.64, 11.17, 1.78, np.nan, np.nan, np.nan]
}
ic_errors = {
    '96 Well Plate': [np.nan]*7,
    'Gradient Plates': [0.12, 3.58, 0.939, 0.207, np.nan, np.nan, np.nan]
}
colors = {
    'Mecillinam': '#440154',
    'Ampicillin': '#21aa8c',
    'Carbenicillin': '#aa2222',
    'Chloramphenicol': '#028bff',
    'Doxycycline': '#0000ff',
    'Erythromycin': '#ff0000',
    'Ciprofloxacin': '#00ff00'
}

# Filter valid data
x_vals, y_vals, x_errs, y_errs, labels, color_vals = [], [], [], [], [], []

for i, ab in enumerate(antibiotics):
    x = mic_data['Gradient Plates'][i]
    y = mic_data['96 Well Plate'][i]
    x_err = ic_errors['Gradient Plates'][i]
    y_err = ic_errors['96 Well Plate'][i]
    if not np.isnan(x) and not np.isnan(y):
        x_vals.append(x)
        y_vals.append(y)
        x_errs.append(x_err if not np.isnan(x_err) else 0)
        y_errs.append(y_err if not np.isnan(y_err) else 0)
        labels.append(ab)
        color_vals.append(colors[ab])

# Plot
fig, ax = plt.subplots(figsize=(8, 6))

# Plot each point with error bars and legend entry
for i in range(len(x_vals)):
    ax.errorbar(
        x_vals[i], y_vals[i],
        xerr=x_errs[i], yerr=y_errs[i],
        fmt='o', markersize=8,
        color=color_vals[i],
        ecolor='gray', capsize=3,
        label=labels[i]
    )

# Remove duplicate labels from legend
handles, unique_labels = ax.get_legend_handles_labels()
by_label = dict(zip(unique_labels, handles))
ax.legend(by_label.values(), by_label.keys(), title="Antibiotic", fontsize=10, title_fontsize=11, loc='lower right')

# Reference line (1:1)
lims = [0.01, 100]
ax.plot(lims, lims, 'k--', lw=1)
ax.set_xlim(lims)
ax.set_ylim(lims)

# Axes and grid
ax.set_title("Correlation of MICs (Gradient vs 96-well)", fontsize=14)
ax.set_xlabel("Gradient Plate MIC (µg/mL)", fontsize=12)
ax.set_ylabel("96 Well Plate MIC (µg/mL)", fontsize=12)
ax.set_xscale('log')
ax.set_yscale('log')
ax.grid(True, which='both', linestyle=':', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

# Prepare lists for plotting
x_vals, y_vals, x_errs, y_errs, labels, color_vals = [], [], [], [], [], []

for i, ab in enumerate(antibiotics):
    x = mic_data['Gradient Plates'][i]
    y = mic_data['96 Well Plate'][i]
    x_err = ic_errors['Gradient Plates'][i]
    y_err = ic_errors['96 Well Plate'][i]

    if not np.isnan(x) and not np.isnan(y):
        x_vals.append(x)
        y_vals.append(y)
        x_errs.append(x_err if not np.isnan(x_err) else 0)
        y_errs.append(y_err if not np.isnan(y_err) else 0)
        labels.append(ab)
        color_vals.append(colors[ab])

# Convert to arrays
x_vals = np.array(x_vals)
y_vals = np.array(y_vals)
x_errs = np.array(x_errs)
y_errs = np.array(y_errs)

# Plot with error bars
ax.errorbar(x_vals, y_vals, xerr=x_errs, yerr=y_errs, fmt='o',
            ecolor='gray', capsize=4, markersize=7,
            markerfacecolor='white', markeredgecolor='black')

# Add colored dots over errorbar markers
for i, (x, y, ab) in enumerate(zip(x_vals, y_vals, labels)):
    ax.plot(x, y, 'o', markersize=7, color=colors[ab], label=ab)

# Annotate
for i, label in enumerate(labels):
    ax.text(x_vals[i]*1.05, y_vals[i], label, fontsize=9, verticalalignment='center')

# 1:1 reference line
lims = [min(x_vals + y_vals)*0.8, max(x_vals + y_vals)*1.2]
ax.plot(lims, lims, 'k--', linewidth=1)
ax.set_xlim(lims)
ax.set_ylim(lims)

# Format
ax.set_title(f"Correlation of MICs (Gradient vs 96-well)", fontsize=14)
ax.set_xlabel("Gradient Plate MIC (µg/mL)", fontsize=12)
ax.set_ylabel("96 Well Plate MIC (µg/mL)", fontsize=12)
ax.set_xscale('log')
ax.set_yscale('log')
ax.grid(True, which='both', linestyle=':', linewidth=0.5)
plt.tight_layout()
plt.show()
```
